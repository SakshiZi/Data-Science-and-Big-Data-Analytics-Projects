```python
#importing necessary libraries
import scipy.stats as stats
import pylab
def normality(data,feature):
    plt.figure(figsize=(15,5))
    plt.subplot(1,2,1)
    sns.kdeplot(data[feature])
    plt.subplot(1,2,2)
    stats.probplot(data[feature],plot=pylab)
    plt.show()
```

```python
normality(df,'price')
```

# Data Transformation
Skewed to normal

# Logarithmic Transformation 
This will convert the Price value to its log value i.e log(Price)

```python
#performing logarithmic transformation on the feature
df['price_log']=np.log(df['price'])
#plotting to check the transformation
normality(df,'price_log')
```

# Reciprocal Transformation 
This will inverse values of Price i.e1/Price

```python
df['price_reciprocal']=1/df.price
normality(df,'price_reciprocal')
```

# Square Root Transformation 
This transformation will take the square root of the Price column i.e sqrt(Price).

```python
df['price_sqroot']=np.sqrt(df.price)
normality(df,'price_sqroot')
```

# Exponential Transformation:
The exponential value of the Price variable will be taken.

```python
df['price_exponential']=df.price**(1/1.2)
normality(df,'price_exponential')
```
