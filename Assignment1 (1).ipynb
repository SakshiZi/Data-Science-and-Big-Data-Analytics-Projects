```python
#1. importing the required libraries
#Panda : Analysing, cleaning, exploring and manipulating data
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#5. Data Formatting
#drop nullues from age
New_train = train.dropna()
New_train.describe()
New_train.head(10)
New_train.shape
```

```python
#Fill null values with mode instead.
x = train['Age'].mode()[0] #0 to specify we want mode of column, 1 is for rows.
train['Age'].fillna(x, inplace = True) #inplace argument makes the change in given dataframe rather than returning a copy
train.isnull().sum()
```

```python
train.describe()
```

```python
#Now convert age into int
train['Age']= train['Age'].astype('int')
train.dtypes
```

```python
train.describe()
```

```python
# 6. Turn Categorical variables into quantitaive variables in python.
train['Sex'].replace(['male', 'female'],[0, 1], inplace=True)
train.head(10)
```

```python
plt.scatter(train['PassengerId'],train['Age'])
```

```python
#Example of Data normalisation
New_train_1 = train
New_train_1['Age'] = ( New_train_1['Age'] - New_train_1['Age'].mean() )/ New_train_1['Age'].std() 
New_train_1.describe()
```
